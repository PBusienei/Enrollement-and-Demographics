```python
#Turn all values to be Numeric
#df = df.apply (pd.to_numeric, errors='coerce')
# Drop the Rows only if the entire row has NAN values
mnps_filtered = mnps_df.dropna(how='all')
```

```python
# Replace Missing values with zero
mnps_filtered2 = mnps_filtered.fillna(0)
```

```python
fig, ax = plt.subplots(figsize=(14,8))

ax.set_ylabel('School Level')

# boxplot with only these two variables
_ = mnps_filtered2[['School Level', 'Limited English Proficiency']].boxplot(by='School Level', figsize=(13,8), vert=False, sym='b.', ax=ax)
```

```python
 # set the size of the graph
plt.figure(figsize=(14,8))
_ = sns.regplot(data=mnps_filtered, x='Limited English Proficiency', y='Economically Disadvantaged')
```

In the scatter plot above, each dot represents a student. The placement of the dot represents that economically disadvantage students(Y-axis) and the percentage of its students on Limited english profiency (x-axis).

The upward trend line shows the positive correlation between Economically disadvantaged students and Limited English profieciency. The slope of the trend line indicates how much family economics plays a role in english proficiency. 

```python
# create tabular correlation matrix
corr = mnps_filtered.corr()
_, ax = plt.subplots(figsize=(13,10)) 

# graph correlation matrix
_ = sns.heatmap(corr, ax=ax,
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values,
                cmap='coolwarm')
```

Red cells indicate positive correlation; blue cells indicate negative correlation; white cells indicate no correlation. The darker the colors, the stronger the correlation (positive or negative) between those two variables.

```python
mnps_filtered.groupby('School Level').Disability.sum().plot(kind='bar')
```

```python
mnps_filtered.groupby('School Name').Disability.sum().plot(kind='bar',figsize=(24,10))
```

```python
mnps_filtered.groupby('School Name')['Economically Disadvantaged'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(24,10))
plt.title('Economically Disadvantaged per School')
plt.xlabel('School Names')
plt.ylabel('Total Number of Students')
```

```python
mnps_filtered.groupby('School Name')['Limited English Proficiency'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(24,10))
plt.title('Limited English Proficiency per School')
plt.xlabel('School Names')
plt.ylabel('Total Number of Students')
```

```python
mnps_filtered.groupby('School Name')['Disability'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(24,10))
plt.title('Disability per School')
plt.xlabel('School Names')
plt.ylabel('Total Number of Students')
```
